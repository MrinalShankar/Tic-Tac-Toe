# Tic-Tac-Toe

```python
t0=True
while(t0):
    def place_mark(board,mark,position):
        board[position]=mark
    
    def print_board(board):
        print(board[7]+' |'+board[8]+' |'+board[9])
        print('--------')
        print(board[4]+' |'+board[5]+' |'+board[6])
        print('--------')
        print(board[1]+' |'+board[2]+' |'+board[3])
        print('___________________________________')
        print()
    

    def win_check(board, mark):
        mark_ind=[]
        w=False
        for i in [1,4,7]:
            if board[i]==mark:
                if (board[i]==board[i+1] and board[i]==board[i+2]):
                    w=True
                    break
   
        for i in [1,2,3]:
            if board[i]==mark:
                if (board[i]==board[i+3] and board[i]==board[i+6]):
                    w=True
                    break
    
        if board[1]==mark:
            if (board[5]==mark and board[9]==mark):
                w=True
            
            
        elif board[3]==mark:
            if (board[5]==mark and board[7]==mark):
                w=True
            
            
        else:
            pass
    
        return w

                 
    def full_board_check(board):
        for x in board:       
                if x ==' ':
                    return False
                else:
                    pass
        return True
   
    def space_check(board, position):
        return (board[position]==' ')

    def player_choice(board):
        t=space_check(board,0)
        while not t:
            print('Enter your next position')
            nxt_pos=int(input())
            t=space_check(board,nxt_pos)
        return nxt_pos


    print('Welcome to Tic-Tac-Toe!')
    mark1=''
    mark2=''
    while(mark1!='X' and mark1!='O'):
        print('Player 1, Enter your choice X or O ')
        mark1=input().upper()
    if mark1=='X':
        mark2='O'
    else:
        mark2='X'
    
    print('Player1 is '+mark1+' and Player 2 is '+mark2)

    test_board=['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
    print_board(test_board)
    t1=False
    t2=False
    t3=False
    t4=False

    while not(t1 or t2 or t3 or t4):
        p1= player_choice(test_board)
        place_mark(test_board,mark1,p1)
        print_board(test_board)
        if win_check(test_board,mark1):
            print('Player 1 wins!!')
            t1=True
            break
    
    
        elif full_board_check(test_board):
            print('Board Full. Game Tied')
            t2=True
            break
        
        else:
            pass

        p2= player_choice(test_board)
        place_mark(test_board,mark2,p2)
        print_board(test_board)
        if win_check(test_board,mark2):
            print('Player 2 wins!!')
            t3=True
            break
    
        elif full_board_check(test_board):
            print('Board Full! Game Tied')
            t4=True
            break
    
        else:
            pass
    print('Game Over')

    print("Do you want to play again Y/N?")
    t=''
    while(t!='Y' and t!='N'):
        t=input().upper()
    if t=='Y':
        t0=True
    else:
        t0=False
        
print("Thanks for playing")
```
